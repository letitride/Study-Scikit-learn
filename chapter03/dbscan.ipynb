### DBSCAN

DBSCANはクラスタに属さないデータポイントも判別できる。

各データポイントは距離esp内にmin_samplesの他データポイントがあるか確認する。

距離esp内のデータポイントがmin_samples数に満たない場合はノイズとなる。

```python
import mglearn
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

小さいデータセットに対してはesp内に納まるmin_samples数のデータポイントが満たない為、すべてノイズと判別される

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs

X, y = make_blobs(random_state=0, n_samples=12)
dbscan = DBSCAN()
cluster = dbscan.fit_predict(X)
print("Cluster memberships:\n{}".format(cluster))
```

```python
mglearn.plots.plot_dbscan()
```

```python
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler

X, y = make_moons(n_samples=200, noise=0.05, random_state=0)
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)

dbscan = DBSCAN()
cluster = dbscan.fit_predict(X_scaled)
plt.scatter(X_scaled[:, 0], X_scaled[:,1], c=cluster, cmap=mglearn.cm2, s=60)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```
